```python
import pandas as pd
from sklearn import preprocessing
from sklearn.cluster import MeanShift, estimate_bandwidth
import numpy as np

label_enc =preprocessing.LabelEncoder()
```

```python

df = pd.read_csv('../data/iris_data.csv', decimal=',')
df.drop(['Petal length', 'Petal width'], inplace=True, axis=1)
df['Species'].unique().tolist()
```

```python

colormap = np.array(['r', 'g', 'b'])

df['Species'] = label_enc.fit_transform(df['Species'].astype(str))
df
df.plot.scatter('Sepal length', 'Sepal width', c=colormap[df['Species']])
```

```python
def mean_shift(data, n_samples=1000):
    bandwidth = estimate_bandwidth(data, quantile=0.2, 
                                   n_samples=n_samples)

    ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
    ms.fit(data)
    labels = ms.labels_
    cluster_centers = ms.cluster_centers_

    labels_unique = np.unique(labels)
    n_clusters = len(labels_unique)

    print('Number of estimated clusters : {}'.format(n_clusters))
    
    return labels, cluster_centers, n_clusters
mean = mean_shift(df)
mean
# print(labels, cluster_centers, n_clusters)
```
